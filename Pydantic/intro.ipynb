```python
from pydantic import BaseModel, Field
from typing import List, Optional
```

```python
class Person(BaseModel):
    name: str
    age: int 
    city:str
person=Person(name="Alice", age=30, city="New York")
print(type(person))
print(person)
```

```python
class Employee(BaseModel):
    employee_id: str
    position: str
    skills: str
    salary: Optional[float] = None
    is_active: bool = True
```

```python
emp1=Employee(
    employee_id="E123",
    position="Software Engineer",
    skills="Python, FastAPI, Pydantic",
    salary=90000.0,
    is_active=True
)
emp1
```

```python
class Classroom(BaseModel):
    room_number: str
    capacity: int
    resources: List[str] = Field(default_factory=list)
    is_available: bool = True
classroom1=Classroom(
    room_number="A101",
    capacity=30,
    resources=["Projector", "Whiteboard", "WiFi"],
    is_available=False
)
classroom1
```

```python
class CMRIT_ID_CARD(BaseModel):
    id : int
    name : str
    ph_no: int
    address: str
try:
    id=CMRIT_ID_CARD(
        id=int(input("Enter the id no: ")),
        name=input("Enter the name of the person"),
        ph_no=int(input("Enter the phone_no: ")),
        address=input("Enter the Address :")
    )
except ValueError as ve:
    print("ValueError:", ve)
```

```python
class Student(BaseModel):
    student_id: str
    name: str
    courses: List[str] = Field(default_factory=list)
    gpa: Optional[float] = None
    is_enrolled: bool = True
student1=Student(
    student_id="S456",
    name="Bob",
    courses=["Math", "Science", "History"],
    gpa=3.8,
    is_enrolled=True
)
# type(student1)
student1
```
